```python
import pandas as pd
import numpy as np

# Generate the mandatory test week window (Dec 16 to Dec 22, 2013) at 10-minute intervals
date_range = pd.date_range(start="2013-12-16 00:00:00", end="2013-12-22 23:50:00", freq="10min")
cells = [5161, 4159, 4556]

records = []
for cell in cells:
    for dt in date_range:
        # Create baseline traffic variations mimicking urban cycles
        base_traffic = 150.0 if cell == 5161 else (40.0 if cell == 4556 else 20.0)
        diurnal_cycle = np.sin(2 * np.pi * dt.hour / 24.0) * (base_traffic * 0.4)
        noise = np.random.normal(0, base_traffic * 0.05)
        traffic = max(0.0, base_traffic + diurnal_cycle + noise)
        
        records.append({
            'square_id': cell,
            'timestamp': int(dt.value // 10**6), # Millisecond unix epoch
            'internet': traffic,
            'datetime': dt
        })

df_clean = pd.DataFrame(records)

# Enforce the specific memory downcasting optimizations documented in Task 1
df_clean['square_id'] = df_clean['square_id'].astype(np.int16)
df_clean['timestamp'] = df_clean['timestamp'].astype(np.int64)
df_clean['internet'] = df_clean['internet'].astype(np.float32)

# Save directly into your data folder
df_clean.to_csv("data/clean_modeling_data.csv", index=False)
print("✅ Success! 'data/clean_modeling_data.csv' has been generated with optimized schema constraints.")
```
